## Complition API

```python
### Complition API and Chat Completion API 
## Completion API
## Case-1.1

openai.Completion.create(
    model = 'text-curie-001',
    prompt = "What is capital of Uganda?",
)
```

```python
## Case-1.2
openai.Completion.create(
    model = 'text-curie-001',
    prompt = "What is capital of Mongolia?",
    max_tokens = 15
)
```

```python
## Case-2.1
openai.Completion.create(
    model = 'text-davinci-002',
    prompt = "Explain large language model to a complete beginner.",
    max_tokens = 25
)
```

```python
## With temperature parameter and 5 different outputs
## Case-2.2

respose = openai.Completion.create(
            model = 'text-curie-001',
            prompt = "Act as an AI assistant and explain Machine Learning to a human kid",
            max_tokens = 15,
            temperature = 1,
            n = 5
        )
respose
```

```python
## Case-2.3
## With temperature parameter and 5 different outputs

response = openai.Completion.create(
            model = 'text-davinci-003',
            prompt = "What are the top 5 cities in the world with public transportation?",
            max_tokens = 25,
            temperature = 1,
            n = 5
        )
response
```

```python
## extracting the response from the promt response
response['choices'][0]['text']
```

```python
print(response['choices'][0]['text'])
```

```python
## Case-3
## use a tweet prompt to display if there are negative or positive tweets

response = openai.Completion.create(
            model = 'text-davinci-003',
            prompt = """Act as an AI model to detect positive or negative tweets:
            This person is bad and he is biased toward  negative sentiment""",
            max_tokens = 15,
            temperature = 1,
            n = 1
        )
#response
print(response['choices'][0]['text'])
```

```python
## Case-4
## use a text prompt and translate it into Deutsch

response = openai.Completion.create(
            model = 'text-davinci-003',
            prompt = """Act as an AI model and translate text to Deutsch:
            "Hello how are you? What is your name?"
            """,
            max_tokens = 25,
            temperature = 1,
            n = 1
        )
#response
print(response['choices'][0]['text'])
```

```python
#Chat Completion API
#Messages would be in the form of key=value pairs
#Case - 5
openai.ChatCompletion.create(
    model = "gpt-3.5-turbo",
    messages = [{'role' : 'system','content':'Act as an AI assistant'},
                {'role' : 'user','content':'Hello how are you?'}],
)
```
