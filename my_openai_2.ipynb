```python
# Example dummy function hard coded to return the same weather
# In production, this could be your backend API or an external API
import requests
def get_current_weather(location):
    """Get the current weather in a given location"""

    url = "https://ai-weather-by-meteosource.p.rapidapi.com/find_places"

    querystring = {"text":location}

    headers = {
      #"X-RapidAPI-Key": "XXX",
      #"X-RapidAPI-Host": "XXX"
      "X-RapidAPI-Key": "815471d58emsh49edc50dc06c76ap1727e7jsna075612f3b50",
      "X-RapidAPI-Host": "ai-weather-by-meteosource.p.rapidapi.com"
    }

    response = requests.get(url, headers=headers, params=querystring)

    print(response.json())
  
    return response.json()
```

```python
response = get_current_weather('Kolkata')
response
```

```python
functions = [
        {
            "name": "get_current_weather",
            "description": "Get the current weather in a given location",
            "parameters": {
                "type": "object",
                "properties": {
                    "location": {
                        "type": "string",
                        "description": "The city and state, e.g. San Francisco, CA",
                    },
                    
                },
                "required": ["location"],
            },
        }
    ]
```

* Sending a message to the chat GPT

```python
user_message = "Hi there"
messages = []
messages.append({"role": "user","content" : user_message})
completion = openai.ChatCompletion.create(
    model = "gpt-3.5-turbo",
    messages = messages
)
```

```python
messages
```

The ChatComplition allows us to append any number of messages to it
, so that it can remem the context
Once the above message is received by the ChatComplition it will answer as below

```python
print(completion.choices[0].message)
```

```python
user_message = "What is the temperature of Kolkata?"
messages.append({"role": "user","content" : user_message})
completion = openai.ChatCompletion.create(
    model = "gpt-3.5-turbo",
    messages = messages,
    functions = functions
)
```

```python
print(completion.choices[0].message)
```

```python
response=completion.choices[0].message
response
```

```python
function_name=response['function_call']['name']
print(function_name)
```

Use 'eval' to cover the string to Json

```python
import json
location = eval(response['function_call']['arguments'])['location']
print(location)
```
